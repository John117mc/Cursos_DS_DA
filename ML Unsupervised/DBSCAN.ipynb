# DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

* Baseado em densidade, agrupando os pontos similares no mesmo espaço
* Não é necessario especificar o número de clusters
* Em geral apresenta melhores resultados do que o Kmeans
* Mais rápido que o kmeans
* Tenta encontrar os pontos que são separados por uma distância não maior do que um limiar (threshold distance)
* Encontra padrões não lineares
* Robusto com outliers
* O resultado pode ser mais consistente que o k-means porque a inicialização dos "centroides" não afeta tanto o algoritmo
* Dependendo da inicialização, um ponto pode pertencer ao cluster dependendo a ordem
* Pode ser díficil encontrar um bom valor para o parâmetro de distância

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

## Credit Card

```python
df_credit = pd.read_csv('../data/credit_card_clients.csv', header=1)
df_credit
```

```python
df_credit['BILL TOTAL'] = df_credit['BILL_AMT1'] + df_credit['BILL_AMT2'] + df_credit['BILL_AMT3'] + df_credit['BILL_AMT4'] + df_credit['BILL_AMT5'] + df_credit['BILL_AMT6']
df_credit 
```

```python
X_credit = df_credit.iloc[:, [1, 25]].values
X_credit
```

```python
scaler_credit = StandardScaler()
X_credit_s = scaler_credit.fit_transform(X_credit)
X_credit_s
```

```python
dbscan_credit = DBSCAN(eps=0.37, min_samples= 5)
label = dbscan_credit.fit_predict(X_credit_s)
label
```

```python
np.unique(label, return_counts=True)
```

```python
sns.set(rc ={'figure.figsize': (15,6)})
sns.scatterplot(x= X_credit_s[:,0], y= X_credit_s[:,1], hue= label, palette= 'viridis');
```

## Salary and Age

```python
df_salary = np.array([[20, 1000],[27,1200],[21,2900],[37,1850],[46,900],[53,950],
                       [55,2000],[47,2100],[52,3000],[32,5900],[39,4100],[41,5100],
                       [39,7000],[48,5000],[48,6500]])
```

```python
scaler_salary = StandardScaler()
df_salary_s = scaler_salary.fit_transform(df_salary)
```

```python
dbscan_salary = DBSCAN(eps=0.95, min_samples= 2)
dbscan_salary.fit(df_salary_s)
label = dbscan_salary.labels_
label
```

```python
sns.set(rc= {'figure.figsize': (15,7)})
sns.scatterplot(x= df_salary_s[:,0], y= df_salary_s[:,1], hue= label, palette= 'viridis');
```
