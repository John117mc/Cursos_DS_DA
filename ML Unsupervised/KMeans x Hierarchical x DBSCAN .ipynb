```python
import seaborn as sns

from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
X_random, y_random = datasets.make_moons(n_samples= 1500, noise=0.09)
```

```python
sns.set(rc= {'figure.figsize': (15,7)})
sns.scatterplot(x= X_random[:,0], y= X_random[:,1], palette= 'viridis');
```

## K-Means

```python
kmeans = KMeans(n_clusters= 2)
label = kmeans.fit_predict(X_random)

sns.scatterplot(x= X_random[:,0], y= X_random[:,1], hue= label, palette= 'viridis');
```

## Hierarchical Clustering

```python
hc = AgglomerativeClustering(n_clusters= 2, linkage= 'ward')
label = hc.fit_predict(X_random)

sns.scatterplot(x= X_random[:,0], y= X_random[:,1], hue= label, palette= 'viridis');
```

## DBSCAN

```python
dbscan = DBSCAN(eps= 0.1)
label = dbscan.fit_predict(X_random)

sns.scatterplot(x= X_random[:,0], y= X_random[:,1], hue= label, palette= 'viridis');
```
