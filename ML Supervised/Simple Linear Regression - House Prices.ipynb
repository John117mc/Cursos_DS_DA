```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns

from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
house = pd.read_csv('../data/house_prices.csv')
house
```

```python
house.describe()
```

```python
house.isnull().sum()
```

```python
fig = plt.figure(figsize=(15,15))
sns.heatmap(house.corr(), annot=True);
```

```python
X_house = house.iloc[:, 5:6].values
y_house = house['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_house, y_house, test_size= 0.3, random_state= 0)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
pred = model.predict(X_train)
```

```python
sns.set_style("darkgrid")

fig , ax = plt.subplots()
fig.set_size_inches(13.0, 8.5)

sns.scatterplot(x= X_train.ravel(), y= y_train, ax=ax)
sns.lineplot(x= X_train.ravel(), y= pred, color='red');
```

```python
pred_test = model.predict(X_test)
```

```python
sns.set_style("darkgrid")

fig , ax = plt.subplots()
fig.set_size_inches(13.0, 8.5)

sns.scatterplot(x= X_test.ravel(), y= y_test, ax=ax)
sns.lineplot(x= X_test.ravel(), y= pred_test, color='red');
```

```python
MAE = mean_absolute_error(y_test, pred_test)
MSE = mean_squared_error(y_test, pred_test)
RMSE = np.sqrt(mean_squared_error(y_test, pred_test))
```
